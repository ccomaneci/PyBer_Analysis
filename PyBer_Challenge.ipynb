```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
city_data_df
```

```python
ride_data_df
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

```python
rides_by_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
rides_by_city_type
```

```python
drivers_by_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
drivers_by_city_type
```

```python
fares_by_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
fares_by_city_type
```

```python
avg_fare_by_city_type = pyber_data_df.groupby(["type"]).mean()["fare"]
avg_fare_by_city_type
```

```python
avg_driver_fare_by_city_type = pyber_data_df.groupby(["type"]).sum()["fare"] / drivers_by_city_type
avg_driver_fare_by_city_type
```

```python
pyber_summary_df = pd.DataFrame({"Total Rides": rides_by_city_type,
                                 "Total Drivers": drivers_by_city_type,
                                 "Total Fares": fares_by_city_type,  
                                 "Average Fare per Ride": avg_fare_by_city_type,
                                 "Average Fare per Driver": avg_driver_fare_by_city_type})
pyber_summary_df
```

```python
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
pyber_summary_df['Total Rides'] = pyber_summary_df['Total Rides'].map('{:,}'.format)
pyber_summary_df['Total Drivers'] = pyber_summary_df['Total Drivers'].map('{:,}'.format)
pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].map('${:,.2f}'.format)
pyber_summary_df['Average Fare per Ride'] = pyber_summary_df['Average Fare per Ride'].map('${:,.2f}'.format)
pyber_summary_df['Average Fare per Driver'] = pyber_summary_df['Average Fare per Driver'].map('${:,.2f}'.format)

pyber_summary_df
```

```python
pyber_data_df
```

```python
fare_by_city_type_date_df = pyber_data_df.groupby(['type','date']).sum().fare.to_frame()
fare_by_city_type_date_df
```

```python
fare_by_city_type_date_df = fare_by_city_type_date_df.reset_index()
fare_by_city_type_date_df
```

```python
fare_by_city_type_date_df = fare_by_city_type_date_df.pivot(index='date',columns='type', values='fare')
fare_by_city_type_date_df
```

```python
fares_by_dates_df = fare_by_city_type_date_df.loc['2019-01-01':'2019-04-28']
fares_by_dates_df
```

```python
fares_by_dates_df.index = pd.to_datetime(fares_by_dates_df.index)
fares_by_dates_df
```

```python
fares_by_dates_df.info()
```

```python
fares_by_dates_df = fares_by_dates_df.resample('W').sum()
fares_by_dates_df
```

```python
from matplotlib import style
style.use('fivethirtyeight')

fares_by_dates_df.plot(figsize=(20,5))
plt.title("Total Fare by City Type")
plt.ylabel('Fare ($USD)')
plt.xlabel('Month')
plt.grid(True)

plt.legend()

plt.show()

plt.savefig('Analysis/Pyber_fare_summary.png')
```
